Así, la primera forma de crear un DataFrame de pandas en Python, es tener un diccionario, donde la clave es el nombre de la columna y el valor debe ser una lista con los valores que vamos agregar a la columna.

```python
data3 = {
    'Equipo': ['A', 'A', 'B', 'B'],
    'Jugador': ['Alice', 'Bob', 'Charlie', 'David'],
    'Puntos': [15, 20, 10, 25],
    'Asistencias': [5, 7, 8, 6]
}
df3 = pd.DataFrame(data3)
df3.head()
```

```python
grupos = df3.groupby('Equipo')

for etiqueta, grupo in grupos:
  print(f'Grupo: {etiqueta}')
  print(grupo)
  print('===============')
```

```python
dataresumen = df3.groupby('Equipo').agg({
    "Puntos": ['mean', 'sum'],
    "Asistencias": "sum"
})

dataresumen
```
